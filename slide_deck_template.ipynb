## Investigation Overview

> In this investigation, I wanted to look at the characteristics of gender, weekday and trip start time that could be used to predict members uses habits,  The main focus was on gender, trip start time, weekday statistics and age groups.

## Dataset Overview

> The data consisted of 183412 row , 16 columns consist of time, numeric and catogrical columns,  The attributes included time, trip duration, long/lat, gender, date of birth and customer type, Gender column, duration time column, week days column and age column are most interinsting for me.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
base_color = sb.color_palette()[0]
# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_clean = pd.read_csv('gofordbike_clean.csv')
```

> Note that the above cells have been set as "Skip"-type slides. That means
that when the notebook is rendered as http slides, those cells won't show up.

## (Week Day VS Duration)

>#### Saturday has longer average trip duration 

```python
plt.figure(figsize=(8,8))
color1 = sb.color_palette("flare")
plt.pie(df_clean.groupby('day_of_week')['duration_hr'].mean(), labels=df_clean.groupby('day_of_week')['duration_hr'].mean().index, colors=color1, startangle=90, autopct='%1.1f%%' , explode=[0.1, 0.1, 0.4, 0.1, 0.1, 0.1, 0.1], shadow=True ) 
plt.legend(title = "Week Days")
plt.title('Week days VS Duration', fontsize=25)
plt.axis('equal');
```

## (Week Day VS Trip Start Hour)

> #### All days have almost the same start hour but the median for friday looks alittle bit earlier, Users start early at fridays and finish late at wensday and thursday.

```python
plt.figure(figsize=(12,8))
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
sb.boxplot(df_clean.day_of_week, df_clean.start_hr, color=base_color, order=day_order)
plt.xlabel('Day_of_week', fontsize=25)
plt.ylabel('Trip Start Hour', fontsize=25)
plt.title('Week Day VS Trip Start Hour', fontsize=25);
```

## (Trip duration vs week day and age and gender )

> #### > Females start early than males.
> 
> #### > Both females and males start trips early at fridays
>
> #### > Older men are starting trips earlier than other age groups
>
> #### > Females trip duration is longer 

```python
fig = plt.figure(figsize=[28, 28])

fig.suptitle('MULTIVARIANTE PLOTS SHOWING RELATION BETWEEN TRIP START HOUR, WEEK DAYS, AGE AND GENDER', fontsize=28,  y = 1 )

plt.subplot(3, 1, 1)
sb.pointplot(data=df_clean, x='start_hr', y='duration_min', hue='member_gender', errwidth=0)
plt.xlabel('Trip Start Hour', fontsize=28)
plt.ylabel('Trip Duration Per Minutes', fontsize=28)
plt.title('TRIP START HOUR VS DURATION PER MINTUNE', fontsize=28)
plt.legend(title='Gender', fontsize=28);

plt.subplot(3, 1, 2)
sb.pointplot(data=df_clean, x='start_hr', y='age', hue='member_gender', errwidth=0)
plt.xlabel('Trip Start Hour', fontsize=28)
plt.ylabel('Age', fontsize=28)
plt.title('TRIP START HOUR VS USER AGE', fontsize=28)
plt.legend(title='Gender', fontsize=28);

plt.subplot(3, 1, 3)
sb.pointplot(data=df_clean, x='day_of_week', y='start_hr', hue='member_gender', errwidth=0, order=day_order)
plt.xlabel('Day_of_week', fontsize=28)
plt.ylabel('Trip Start Hour', fontsize=28)
plt.title('Week Day VS Trip Start Hour', fontsize=28)
plt.legend(title='Gender', fontsize=28);
```
